1. What is a random variable in probability theory?

   - A random variable is a numerical outcome of a random phenomenon.

2. What are the types of random variables?

   - The two types are discrete random variables and continuous random variables.

3. What is the difference between discrete and continuous distributions?

   - Discrete distributions involve countable outcomes, while continuous distributions involve uncountably infinite outcomes over an interval.

4. What are probability distribution functions (PDF)?

   - PDF describes the likelihood of a random variable taking a specific value (for discrete) or range (for continuous).

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

   - CDF gives the probability that a random variable is less than or equal to a certain value, while PDF gives the probability at a specific point or density.

6. What is a discrete uniform distribution?

   - It is a distribution where all outcomes are equally likely over a finite set of values.

7. What are the key properties of a Bernoulli distribution?

   - It models a single trial with two outcomes: success (1) with probability p and failure (0) with probability 1-p.

8. What is the binomial distribution, and how is it used in probability?

   - It models the number of successes in a fixed number of independent Bernoulli trials with the same probability of success.

9. What is the Poisson distribution and where is it applied?

   - It models the number of events occurring in a fixed interval of time or space when events occur independently at a constant average rate.

10. What is a continuous uniform distribution?

    - It describes a situation where all values in a continuous range are equally likely.

11. What are the characteristics of a normal distribution?

    - It is symmetric, bell-shaped, and fully defined by its mean and standard deviation.

12. What is the standard normal distribution, and why is it important?

    - It is a normal distribution with mean 0 and standard deviation 1, used for standardizing data and computing probabilities.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

    - CLT states that the sampling distribution of the sample mean approaches a normal distribution as sample size increases, regardless of the population's distribution.

14. How does the Central Limit Theorem relate to the normal distribution?

    - It explains why normal distribution appears frequently in statistics when averaging large samples.

15. What is the application of Z statistics in hypothesis testing?

    - Z statistics are used to determine how far a sample mean deviates from the population mean under the null hypothesis.

16. How do you calculate a Z-score, and what does it represent?

    - Z = (X - μ) / σ; it represents how many standard deviations a value is from the mean.

17. What are point estimates and interval estimates in statistics?

    - A point estimate gives a single value as an estimate, while an interval estimate provides a range of values.

18. What is the significance of confidence intervals in statistical analysis?

    - They provide a range within which the true population parameter is likely to lie with a certain probability.

19. What is the relationship between a Z-score and a confidence interval?

    - Z-scores determine the margin of error used to construct confidence intervals.

20. How are Z-scores used to compare different distributions?

    - Z-scores standardize values, allowing comparison across distributions with different means and variances.

21. What are the assumptions for applying the Central Limit Theorem?

    - Samples must be independent, identically distributed, and have a finite variance; large sample sizes improve accuracy.

22. What is the concept of expected value in a probability distribution?

    - It is the long-run average or mean value of the random variable, weighted by probabilities.

23. How does a probability distribution relate to the expected outcome of a random variable?

    - The probability distribution defines the likelihood of each outcome, which determines the expected value.

```python
# 2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters
low, high = 1, 7  # Simulate a dice
x = np.arange(low, high)

# PMF
pmf_vals = randint.pmf(x, low, high)

# Plot
plt.stem(x, pmf_vals)
plt.title("Discrete Uniform Distribution PMF")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

from scipy.stats import bernoulli

def bernoulli_pdf(p):
    x = [0, 1]
    pdf_vals = bernoulli.pmf(x, p)
    return dict(zip(x, pdf_vals))

# Example
print("Bernoulli PDF (p=0.5):", bernoulli_pdf(0.5))
```

```python
# 4. Write a Python script to simulate a binomial distribution with n=10 and p=O.5, then plot its histogram.

from scipy.stats import binom
import matplotlib.pyplot as plt

n, p = 10, 0.5
data = binom.rvs(n, p, size=1000)

plt.hist(data, bins=range(12), density=True, alpha=0.7, edgecolor='black')
plt.title("Binomial Distribution Histogram (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# 5. Create a Poisson distribution and visualize it using Python.

from scipy.stats import poisson

mu = 4
x = np.arange(0, 15)
pmf_vals = poisson.pmf(x, mu)

plt.stem(x, pmf_vals)
plt.title("Poisson Distribution (λ=4)")
plt.xlabel("k")
plt.ylabel("P(X=k)")
plt.grid(True)
plt.show()
```

```python
# 6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

from scipy.stats import randint

low, high = 1, 7
x = np.arange(low, high)
cdf_vals = randint.cdf(x, low, high)

plt.step(x, cdf_vals, where='mid')
plt.title("CDF - Discrete Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.grid(True)
plt.show()
```

```python
# 10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import seaborn as sns

# Simulate a non-normal (exponential) distribution
population = np.random.exponential(scale=2, size=100000)

# Take many samples and compute sample means
sample_means = []
for _ in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(np.mean(sample))

# Plot
sns.histplot(sample_means, kde=True, bins=30, color='skyblue')
plt.title("Central Limit Theorem Demo (Sample Means from Exponential Distribution)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# 11. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sample_means = []

for _ in range(1000):
    sample = np.random.normal(loc=50, scale=10, size=30)
    sample_means.append(np.mean(sample))

sns.histplot(sample_means, kde=True)
plt.title("Central Limit Theorem Verification")
plt.xlabel("Sample Means")
plt.grid(True)
plt.show()
```

```python
# 12. Write a Python function to calculate and plot the standard normal distribution (mean = O, std = 1).

from scipy.stats import norm

def plot_standard_normal():
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)
    plt.plot(x, y)
    plt.title("Standard Normal Distribution (μ=0, σ=1)")
    plt.xlabel("x")
    plt.ylabel("PDF")
    plt.grid(True)
    plt.show()

plot_standard_normal()
```

```python
# 14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

def calculate_z_score(x, mean, std):
    z = (x - mean) / std
    print(f"Z-score: {z}")
    return z

calculate_z_score(70, mean=50, std=10)  # Example
```

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset.

import scipy.stats as stats

sample = np.random.normal(loc=52, scale=10, size=30)
pop_mean = 50
sample_mean = np.mean(sample)
std_error = np.std(sample, ddof=1) / np.sqrt(len(sample))
z = (sample_mean - pop_mean) / std_error
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Z = {z:.2f}, p-value = {p_value:.4f}")
```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result.

sample = np.random.normal(loc=100, scale=15, size=50)
mean = np.mean(sample)
se = stats.sem(sample)
ci = stats.norm.interval(0.95, loc=mean, scale=se)

print(f"95% Confidence Interval: {ci}")
```

```python
# 17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

data = np.random.normal(75, 10, 100)
mean = np.mean(data)
se = stats.sem(data)
ci = stats.norm.interval(0.95, loc=mean, scale=se)

print(f"Mean: {mean:.2f}, 95% CI: {ci}")
```

```python
# 19. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

from scipy.stats import poisson

mu = 3
x = np.arange(0, 10)
cdf_vals = poisson.cdf(x, mu)

for val, cdf in zip(x, cdf_vals):
    print(f"CDF P(X<={val}) = {cdf:.4f}")
```

```python
# 21. Write a Python program to compare the standard deviations of two datasets and visualize the difference.

data1 = np.random.normal(50, 5, 1000)
data2 = np.random.normal(50, 10, 1000)

std1, std2 = np.std(data1), np.std(data2)
print(f"Std Dev 1: {std1:.2f}, Std Dev 2: {std2:.2f}")

plt.hist(data1, bins=30, alpha=0.6, label='σ=5')
plt.hist(data2, bins=30, alpha=0.6, label='σ=10')
plt.legend()
plt.title("Comparison of Standard Deviations")
plt.grid(True)
plt.show()
```

```python
# 22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

data = np.random.normal(100, 15, 1000)
data_range = np.ptp(data)  # Peak to peak (max - min)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print(f"Range: {data_range:.2f}, IQR: {iqr:.2f}")
```

```python
# 23. Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Generate synthetic data
data = np.random.normal(50, 20, 1000)

# Apply Z-score normalization
z_scores = zscore(data)

# Plot original and normalized data
plt.hist(data, bins=30, alpha=0.5, label='Original Data')
plt.hist(z_scores, bins=30, alpha=0.5, label='Z-score Normalized')
plt.title("Z-score Normalization")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```
